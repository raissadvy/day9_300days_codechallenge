```python
import pandas as pd 
from scipy import stats
import seaborn as sns
import numpy as np
import sklearn.preprocessing as preproc
```

```python
df = pd.read_csv('/Users/ndaru/Documents/rai_pwdk/self_learn/daily_challenge/#6_day/insurance.csv', low_memory=False)
df.head()
```

```python
df.describe()
```

```python
# Min-max scaling
df['minmax'] = preproc.minmax_scale(df['charges'])
```

```python
sns.histplot(df['charges'])
```

```python
sns.histplot(df['minmax'])
```

```python
# Standard scaling
y = df['charges']
y = np.array(y).reshape(-1,1)
df['standardized'] = preproc.StandardScaler().fit_transform(y)
```

```python
sns.histplot(df['standardized'])
```

```python
# L2 normalization
df['l2_normalized'] = preproc.normalize(y, axis=0)
```

```python
sns.histplot(df['l2_normalized'])
```

```python
# Robustscaler
df['robust_scaler'] = preproc.RobustScaler().fit_transform(y)
```

```python
sns.histplot(df['robust_scaler'])
```

```python
# Quantile transform
df['quantile_transform'] = preproc.quantile_transform(y, n_quantiles=10, random_state=0, copy=True)
```

```python
sns.histplot(df['quantile_transform'])
```

```python
# Max abs scaler
df['max_abs_scaler'] = preproc.MaxAbsScaler().fit_transform(y)
```

```python
sns.histplot(df['max_abs_scaler'])
```

```python
# Normalize
df['normalize'] = preproc.normalize(y, norm='l1', axis=1, copy=True, return_norm=False)
```

```python
sns.histplot(df['normalize'])
```

```python
df['normalize2'] = preproc.normalize(y, norm='l2', axis=1, copy=True, return_norm=False)
sns.histplot(df['normalize2'])
```

```python
df['normalizeMax'] = preproc.normalize(y, norm='max', axis=1, copy=True, return_norm=False)
sns.histplot(df['normalizeMax'])
```
